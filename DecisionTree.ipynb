```python
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.metrics import confusion_matrix, accuracy_score, mean_absolute_error, mean_squared_error
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
x_c, y_c = datasets.load_iris(return_X_y = True)
x_r, y_r = datasets.fetch_california_housing(return_X_y=True)
```

```python
x_c_train, x_c_test, y_c_train, y_c_test = train_test_split(x_c, y_c, test_size=0.2, random_state=56, stratify=y_c)
```

```python
clf_tree_entropy = DecisionTreeClassifier(criterion='entropy')
clf_tree_entropy.fit(x_c_train, y_c_train)

print(accuracy_score(y_c_test, clf_tree_entropy.predict(x_c_test)))
print(confusion_matrix(y_c_test, clf_tree_entropy.predict(x_c_test)))
```

```python
plt.figure(figsize=(15,15))
tree.plot_tree(clf_tree_entropy, filled=True)
```

```python
clf_tree_gini = DecisionTreeClassifier(criterion='gini')
clf_tree_gini.fit(x_c_train, y_c_train)

print(accuracy_score(y_c_test, clf_tree_gini.predict(x_c_test)))
print(confusion_matrix(y_c_test, clf_tree_gini.predict(x_c_test)))
```

```python
plt.figure(figsize=(15,15))
tree.plot_tree(clf_tree_gini, filled=True)
```

# Decision Tree Regressor

```python
ss = StandardScaler()
ss.fit(x_r)

x_r = ss.transform(x_r)
```

```python
x_r_train, x_r_test, y_r_train, y_r_test = train_test_split(x_r, y_r, test_size=0.2, random_state=56)
```

```python
reg_tree = DecisionTreeRegressor(criterion='mse')
reg_tree.fit(x_r_train, y_r_train)

print(mean_squared_error(y_r_test, reg_tree.predict(x_r_test)))
print(mean_absolute_error(y_r_test, reg_tree.predict(x_r_test)))
```
