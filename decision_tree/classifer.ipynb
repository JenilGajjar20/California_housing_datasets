# Decision Tree Classifier

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, mean_absolute_error, mean_squared_error
from sklearn import tree
from matplotlib import pyplot as plt
```

```python
x_c, y_c = datasets.load_iris(return_X_y=True)
x_r, y_r = datasets.fetch_california_housing(return_X_y=True)
```

```python
x_c_train, x_c_test, y_c_train, y_c_test = train_test_split(x_c, y_c, test_size=0.2, random_state=50, stratify=y_c)
```

```python
clf_tree_entropy = DecisionTreeClassifier(criterion='entropy')
clf_tree_entropy.fit(x_c_train, y_c_train)

print(accuracy_score(y_c_test, clf_tree_entropy.predict(x_c_test)))
print(confusion_matrix(y_c_test, clf_tree_entropy.predict(x_c_test)))
```

```python
plt.figure(figsize=(15,15))
tree.plot_tree(clf_tree_entropy, filled=True)
```

```python
clf_tree_gini = DecisionTreeClassifier(criterion='gini')
clf_tree_gini.fit(x_c_train, y_c_train)

print(accuracy_score(y_c_test, clf_tree_gini.predict(x_c_test)))
print(confusion_matrix(y_c_test, clf_tree_gini.predict(x_c_test)))
```

```python
plt.figure(figsize=(15,15))
tree.plot_tree(clf_tree_gini, filled=True)
```
