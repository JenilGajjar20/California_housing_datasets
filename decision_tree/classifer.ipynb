# Decision Tree Classifier

### Importing Libraries

- **datasets**: Import functions to load datasets, including Iris dataset.
- **train_test_split**: Import a function to split the dataset into training and testing sets.
- **DecisionTreeClassifier**: Import the decision tree classifier from scikit-learn.
- **confusion_matrix** and **accuracy_score**: Import evaluation metrics for the classifier.
- **tree**: Import functions for visualizing decision trees.
- **pyplot** from **matplotlib**: Import a module for creating plots.

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn import tree
from matplotlib import pyplot as plt
```

### Loading Datasets

- **datasets.load_iris()**: Load the dataset, which is commonly used for classification tasks. 
    - **x_c** contains the feature data and **y_c** contains the target labels.

```python
x_c, y_c = datasets.load_iris(return_X_y=True)
```

### Splitting the Dataset

- Split the dataset into training and testing sets. The data is split with 80% for training (**x_c_train** and **y_c_train**) and 20% for testing (**x_c_test** and **y_c_test**).
- **test_size=0.2** specifies that 20% of the data should be used for testing
- **random_state=0.5** is the seed for the random number generator to ensure reproducibility
- **stratify=y_c** ensures that the class distribution in both the training and testing sets is similar to the original dataset.

```python
x_c_train, x_c_test, y_c_train, y_c_test = train_test_split(x_c, y_c, test_size=0.2, random_state=50, stratify=y_c)
```

### Decision Tree with Entropy Criterion

- Create a decision tree classifier with the "entropy" criterion and fit it to the training data.
- The **clf_tree_entropy** object now represents the trained decision tree model.

### Model Evaluation (Entropy)

- Calculate and print the accuracy score and confusion matrix for the model trained with the entropy criterion on the test data.

```python
clf_tree_entropy = DecisionTreeClassifier(criterion='entropy')
clf_tree_entropy.fit(x_c_train, y_c_train)

print(accuracy_score(y_c_test, clf_tree_entropy.predict(x_c_test)))
print(confusion_matrix(y_c_test, clf_tree_entropy.predict(x_c_test)))
```

### Visualize Decision Tree (Entropy)

- Create a large figure and visualize the decision tree model trained with the entropy criterion using **tree.plot_tree()**. 
- The **filled=True** parameter fills the nodes with colors to indicate the majority class.

```python
plt.figure(figsize=(15,15))
tree.plot_tree(clf_tree_entropy, filled=True)
```

### Decision Tree with Gini Criterion

- Create a decision tree classifier with the "gini" criterion and fit it to the training data.
- The **clf_tree_gini** object now represents the trained decision tree model.

### Model Evaluation (Gini)

- Calculate and print the accuracy score and confusion matrix for the model trained with the gini criterion on the test data.

```python
clf_tree_gini = DecisionTreeClassifier(criterion='gini')
clf_tree_gini.fit(x_c_train, y_c_train)

print(accuracy_score(y_c_test, clf_tree_gini.predict(x_c_test)))
print(confusion_matrix(y_c_test, clf_tree_gini.predict(x_c_test)))
```

### Visualize Decision Tree (Gini)

- Create a large figure and visualize the decision tree model trained with the gini criterion using **tree.plot_tree()**. 
- The **filled=True** parameter fills the nodes with colors to indicate the majority class.

```python
plt.figure(figsize=(15,15))
tree.plot_tree(clf_tree_gini, filled=True)
```
